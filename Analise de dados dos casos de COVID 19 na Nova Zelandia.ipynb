# Analise de dados dos casos de COVID 19 na Nova Zelandia
Os dados analisados são entre o perido de 03-22-2020 à 07-03-2020.

E foram obtidos em https://github.com/CSSEGISandData/COVID-19

```python
# Importanto bibliotecas.
import pandas as pd
from datetime import date
from datetime import timedelta
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configurações gerais para os graficos.
sns.set(rc={"figure.figsize": (15, 8)})
sns.set(style="whitegrid")
```

```python
# Configuração das variaveis para leitura dos arquivos.
path = 'csse_covid_19_daily_reports/'
extension = '.csv'
filenames = []
d = date(2020, 3, 21)
for x in range(104):
    d = d + timedelta(days = 1)
    date_text = d.strftime("%m-%d-%Y")
    filenames.append(date_text)
```

```python
# Função para percorrer os arquivos de dados e popular o dataframe.
def populate_data_frame():
    dfList = list()
    column = 'Country_Region == "New Zealand"'

    for file in filenames:
        df= pd.read_csv(path+file+extension)
        dfList.append(df.query(column))

    df = pd.concat(dfList)

    return df
```

```python
# Quantidade de linhas e colunas no dataframe.
df = populate_data_frame()
df.reset_index()
df.shape
```

```python
# Apagando as colunas que não seram utilizadas.
df.drop('FIPS', inplace=True, axis=1)
df.drop('Admin2', inplace=True, axis=1)
df.drop('Province_State', inplace=True, axis=1)
df.drop('Lat', inplace=True, axis=1)
df.drop('Long_', inplace=True, axis=1)
df.drop('Combined_Key', inplace=True, axis=1)
df.drop('Last_Update', inplace=True, axis=1)
```

```python
# Exemplo dos dados que seram utilizados.
df.head(10)
```

# Dados gerais
## 1- Índice de casos fatais:

```python
values = np.zeros((104, 1), dtype=np.float64)
dates = pd.date_range("22 3 2020", periods=104, freq="D")

incidence_rate = []
case_fatality_ratio = []

for index, row in df.iterrows():
    incidence_rate.append(row["Case-Fatality_Ratio"])

for i in range(104):
    values[i][0] = incidence_rate[i]

data = pd.DataFrame(values, dates, columns=["Case-Fatality_Ratio"])
data = data.rolling(7).mean()

sns.lineplot(data=data, palette="tab10", linewidth=2.5)
```

## 2 - Índice de ocorrências:

```python
values = np.zeros((104, 1), dtype=np.float64)
dates = pd.date_range("22 3 2020", periods=104, freq="D")

incidence_rate = []
case_fatality_ratio = []

for index, row in df.iterrows():
    incidence_rate.append(row["Incidence_Rate"])

for i in range(104):
    values[i][0] = incidence_rate[i]

data = pd.DataFrame(values, dates, columns=["Incidence_Rate"])
data = data.rolling(7).mean()

sns.lineplot(data=data, palette="tab10", linewidth=2.5)
```

## 3 - Números gerais:

```python
values = np.zeros((104, 4), dtype=np.float64)
dates = pd.date_range("22 3 2020", periods=104, freq="D")

deaths = []
confirmeds = []
actives = []
recovereds = []

for index, row in df.iterrows():
    deaths.append(row["Deaths"])
    confirmeds.append(row["Confirmed"])
    recovereds.append(row["Recovered"])
    actives.append(row["Active"])

for i in range(104):
    values[i][0] = confirmeds[i]
    values[i][1] = recovereds[i]
    values[i][2] = actives[i]
    values[i][3] = deaths[i]
    
data = pd.DataFrame(values, dates, columns=["Confirmed", "Recovered", "Active", "Deaths"])
data = data.rolling(7).mean()

sns.lineplot(data=data, palette="tab10", linewidth=2.5)
```
